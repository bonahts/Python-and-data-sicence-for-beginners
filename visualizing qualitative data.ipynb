```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
listings = pd.read_csv('listings.csv')
listings
```

```python
sns.countplot(x = 'neighbourhood_group', data = listings)
plt.show()
```

countplot merupakan method yang terdapat di library seaborn

```python
sns.barplot(x = 'neighbourhood_group', y = 'minimum_nights', data = listings)
plt.show()
```

penerapan barplot

```python
sns.barplot(x = 'neighbourhood_group', y = 'minimum_nights', data = listings, ci=False)
plt.show()
```
